6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'


```python
marks=float(input("enter the marks "))
if (marks>90):
    print('A')
elif (marks <90 and marks>=75):
    print('B')
elif (marks > 75 and marks >=50):
    print('C')
else:
    print("fail")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Rakesh","Teja","Sai","Kiran","Mahesh","Rakesh","Anil","Sunil","Ravi","Kiran"],
    "Age": [22,23,24,25,26,22,23,24,25,25],
    "City": ["Hyd","Blr","Hyd","Chennai","Hyd","Hyd","Blr","Hyd","Pune","Chennai"],
    "Salary": [30000,40000,np.nan,35000,45000,30000,40000,38000,np.nan,35000]
})

```
